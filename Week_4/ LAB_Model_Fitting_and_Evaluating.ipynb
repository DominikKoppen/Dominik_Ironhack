```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PowerTransformer, StandardScaler, MinMaxScaler, OrdinalEncoder
```

```python
data = pd.read_csv('marketing_customer_analysis_clean.csv')
```

```python
data
```

```python
data_numericals = data.select_dtypes(include="number")
```

```python
X = data_numericals.drop("total_claim_amount", axis=1)
```

```python
data_numericals
```

```python
X
```

```python

y = data[['total_claim_amount']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 100)
```

```python
#Standardize the data (after the data split!
```

```python
scaler = StandardScaler()
```

```python
scaler = scaler.fit(X_train)
```

```python
scaled_data_X_train = scaler.transform(X_train)
```

```python
scaled_data_X_train
```

```python
scaled_data_X_test = scaler.transform(X_test)
scaled_data_X_test
```

```python
scaled_data_y_train = scaler.transform(y_train)
scaled_data_y_test = scaler.transform(y_test)
```

```python
scaled_data_y_train
```

```python
scaled_data_y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
model = lm.fit(scaled_data_X_train, y_train)
```

```python
# here i am testing the data to see if it works
```

```python
y_pred = model.predict(scaled_data_X_test)
```
