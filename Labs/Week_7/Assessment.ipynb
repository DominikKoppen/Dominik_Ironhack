```python
def feature_engineering(X_train, X_test, label_encoders={'Sex': LabelEncoder(), 'Embarked': OneHotEncoder()}, 
                        drop_cols=['SibSp', 'Parch', 'Sex', 'Embarked', 'PassengerId', 'Name', 'Ticket']):
   
    for col, encoder in label_encoders.items():
        if isinstance(encoder, LabelEncoder):
            X_train[f'{col}_Code'] = encoder.fit_transform(X_train[col])
            X_test[f'{col}_Code'] = encoder.transform(X_test[col])
        elif isinstance(encoder, OneHotEncoder):
            encoded_cols = pd.DataFrame(encoder.fit_transform(X_train[[col]]).toarray(), 
                                        columns=encoder.get_feature_names_out([col]))
            X_train = pd.concat([X_train, encoded_cols], axis=1)
            
            encoded_cols = pd.DataFrame(encoder.transform(X_test[[col]]).toarray(), 
                                        columns=encoder.get_feature_names_out([col]))
            X_test = pd.concat([X_test, encoded_cols], axis=1)
    
    
    X_train['TravelAlone'] = (X_train['SibSp'] + X_train['Parch'] == 0).astype(int)
    X_test['TravelAlone'] = (X_test['SibSp'] + X_test['Parch'] == 0).astype(int)
    
    
    X_train = X_train.reset_index(drop=True)
    X_test = X_test.reset_index(drop=True)
    
    
    X_train = X_train.drop(columns=drop_cols, axis=1)
    X_test = X_test.drop(columns=drop_cols, axis=1)
    
    return X_train, X_test
```
